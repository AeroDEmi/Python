# Data Quality Report

```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/Emi/Downloads/denue_00_54_csv/conjunto_de_datos/denue_inegi_54_.csv')
```

```python
columns = pd.DataFrame(list(data.columns.values))
columns
```

```python
data_types = pd.DataFrame(data.dtypes,
                          columns=['Data Type'])
data_types
```

```python
missing_data_counts = pd.DataFrame(data.isnull().sum(),
                                   columns=['Missing Values'])
missing_data_counts
```

```python
present_data_counts = pd.DataFrame(data.count(),
                                   columns=['Present Values'])
present_data_counts
```

```python
percentage_value = pd.DataFrame( columns=['Percentage of missing data'])
for v in list(data.columns.values):
    percentage_value.loc[v]=[data[v].isnull().sum()/len(data.index)]
percentage_value
```

```python
unique_value_counts = pd.DataFrame(columns=['Unique Values'])
for v in list(data.columns.values):
    unique_value_counts.loc[v] = [data[v].nunique()]
unique_value_counts
```

```python
minimum_values = pd.DataFrame(columns=['Minimum Value'])
for v in list(data.columns.values):
    try:
        minimum_values.loc[v] = [data[v].min()]
    except:
        pass
minimum_values
```

```python
maximum_values = pd.DataFrame(columns=['Maximum Value'])
for v in list(data.columns.values):
    try:
        maximum_values.loc[v] = [data[v].max()]
    except:
        pass
maximum_values
```

```python
data_quality_report = data_types.join(present_data_counts).join(missing_data_counts).join(percentage_value).join(unique_value_counts).join(minimum_values).join(maximum_values)
```

```python
print("\nData Quality Report")
print("Total records: {}".format(len(data.index)))
data_quality_report.style.format({
    'Percentage of missing data': '{:,.2%}'.format,
})
```

La confiabilidad del anlisis de datos depende de la calidad de datos que se tengan. Por eso se crean los data quality report, con el objetivo de tener medidas objetivas en la calidad de los datos.
