```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sym
sns.set()
%matplotlib inline
sym.init_printing(use_latex='mathjax')
```

BanBajio ofrece una tasa anual del 8%, sin embargo es una tasa de mínimo de 3 meses, el primer mes obtenemos una tasa del 2.5%, el segundo mes 5% y desde el tercer mes el 8%. El monto minimo es de $\$$10,000, mientra que el máximo monto de inversión es de $\$$300,000. ¿Cómo podría obtener $\$$50,000 en 1 año?

```python
#Capital inicial
C=10000
```

```python
#Nuestra función del primer al segundo mes
y1 = C*(1+0.025/12)
```

```python
#Nuestra función del segundo mes
y2 = y1*(1+0.05/12)
```

```python
#Nuestra función desde el mes 3
def y3(x):
    return y2*(1+0.08/12)**x
```

Para saber cuanto necesitamos invertir para al final tener $\$$50000 tenemos que resolver la ecuación

```python
x=0
sym.var('x')
f=x*(1+0.025/12)*(1+0.05/12)*(1+0.08/12)**15-50000
```

```python
x_c = sym.solve(f, x)
x_c
```

Esto quiere decir que nuestro capital inicial es 44975.4

```python
C=np.sum(x_c)
y1 = C*(1+0.025/12)
y2 = y1*(1+0.05/12)
```
