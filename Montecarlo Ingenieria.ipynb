```python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
```

```python
a=1
U=30
def F(E):
    return a*E/(E+U)**3
```

```python
x=np.linspace(0,350,1000)
b=F(x)
max=max(b)
```

```python

x=np.linspace(0,350,1000)
plt.figure(figsize=(8,8))
plt.plot(x,F(x))
plt.show()
```

```python
U=30
a=1

fig, ax = plt.subplots(figsize = (8,8))
puntos=1000
dentro = 0
for i in range(puntos):
    x = random.uniform(0,350)
    y = random.uniform(0,max)
    if y<F(x) :
        dentro += 1
        plt.scatter(x,y)
    else:
        plt.scatter(x,y,c='k')
plt.ylim(0,max)
plt.show()
```

```python
def Y(E):
    U=4.363 #energia de vaporizacion del titanio en eV
    alpha=.25 #de la grafica de alpha pagina 68
    Mi=6.6335209e-26 #Masa del argon Kg
    Mt=7.9485017e-26 #Masa del titanio Kg
    Zi=18 # Numero atomico del argon
    Zt=22 # Numero atomico del titanio
    Eth=20 # Energia de
    return (1.8*alpha/U)*(Zi*Zt*Mi/Mt*E)**(0.5)/((Zi**(2/3)+Zt**(2/3))**(0.5)*(1+Mt/Mi)**(0.5))*(1-(Eth/E)**(0.5))**2
```

```python
yi=0
Adherido=0
plt.figure(figsize=(12,8))
for v in range(50):
    xi=np.random.uniform(7.5,12.5)
    yi=0
    xx = [xi]
    yy = [yi]
    Angulo1= np.random.uniform(0,np.pi)
    x = xi + 1.2*np.cos(Angulo1)
    y = yi + 1.2*np.sin(Angulo1)
    xx.append(x)
    yy.append(y)
    #plt.scatter(x,y)
    while 0<x and x<22 and 0<y and y<20:
        prob = np.random.random()
        Angulo2= np.random.uniform(0,2*np.pi)
        if prob <= 0.36:
            x = x + 1.2*np.cos(Angulo2)
            y = y + 1.2*np.sin(Angulo2)
        else:
            x = x + 1.2*np.cos(Angulo1)
            y = y + 1.2*np.sin(Angulo1)
        xx.append(x)
        yy.append(y)
        #plt.scatter(x,y)
        if y>=20 and 7.5<=x and x <= 12.5:
            Adherido += 1
            plt.plot(xx,yy,'-o',c='green',lw=0.5)
            break
        else:
            plt.plot(xx,yy,'-o',c='r',lw=0.1)
    #plt.plot(xx,yy,'o-')
plt.xlim(0,20)
plt.ylim(-0.5,20)

currentAxis = plt.gca()
currentAxis.add_patch(patches.Rectangle((7.5,-0.5 ), 5, 0.5,alpha=1,color='k'))
currentAxis.add_patch(patches.Rectangle((7.5,19.8 ), 5, 0.2,alpha=1,color='k'))


plt.show()
print("Las partículas que se adhirieron fueron:{} de {}".format(Adherido, 50))
Ar=50/Y(350)
print("Los iones de argón que golpearon el blanco fueron: {}".format(round(Ar,1)))
```
